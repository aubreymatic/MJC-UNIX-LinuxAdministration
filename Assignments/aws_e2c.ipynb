# What
Create your own server for hosting html, javascript, and etc. with Amazon Web Services (EC2)

# What are we using? What do you need?
- An active AWS account. First time sign-ups are eligible for the [free tier](http://aws.amazon.com/free/) for a year
- With AWS we will use the stock Ubuntu Server.

# How
## AWS Console and EC2 Instance Launch.
- Head over to the [AWS Console](http://aws.amazon.com/console/) and log in.
- Click on EC2
- After clicking on EC2, it should look like this, but could be differnt.
![Imgur](http://i.imgur.com/H8ixyYN.png)

- Now click on the Launch Instance button under Create instance.
- It should take you too this page on choosing your servers operating system.

![Imgur](http://i.imgur.com/6KWzJ5t.png)

- We're going to use the ubuntu server, which is eligible for a free tier.
- You should now see a page where it will let you choose the amount of hardware you need.
- You can use the one that is marked for free tier usage unless you need more hardware performance, but it will not be under the free tier.

![Imgur](http://i.imgur.com/k2mN9oE.png)

- Click the Next: Configure Instance Details button.
- It should take you to a page that looks like this.
- You don't need to configure anything here unless you need multiple of the same instances, but again that won't be under the free tier.

![Imgur](http://i.imgur.com/NZmwqSi.png)

- Click on the Next: Add Storage button.

- It should now take you to a page that looks like this.

![Imgur](http://i.imgur.com/hwGfoUF.png)

- As you can see in the blue box, the free tier can go up to 30 GB of Magnetic Storage-HDD or Solid State Storage-SSD.
- You can set it to 30 GB, but you can also leave it as is if you don't need much storage.
- Click on the Next: Tag Instance button.
- It should take you to a page that looks like this.

![Imgur](http://i.imgur.com/EH8n9Fg.png)

- I never really found a good use for this, so you can just leave it alone.
- Click on the Next: Configure Security Group button.
- It should take you to a page that looks like this.

![Imgur](http://i.imgur.com/W7zIghL.png)

- It will give you a Warning about "Rules with source of 0.0.0.0/0 allow all IP addresses to access your instance. We recommend setting security group rules to allow access from known IP addresses only".
- I recomend you to leave it as is because if you set it to known ip addresses, you will need to set a ip for any new location you are at, so you can connect to the server.
- Note that it will create a new security group. if you ever create another instance, you can use the same security groups.
- Click on the Review and Launch button.
- It will take you to a page that looks like this.

![Imgur](http://i.imgur.com/vrtSkoq.png)

![Imgur](http://i.imgur.com/5Mqg2qQ.png)

- It will ask you to create a new key pair or use an existing one.
- These key pairs are what you use to connect to your server.
- I recomend you to save your key pair to a safe place because you will not be able to download them again.
- You will see this page.

![Imgur](http://i.imgur.com/hdstvuq.png)

- Click on the View Instance button
- You will see this page.

![Imgur](http://i.imgur.com/przSfTg.png)

- Your instance should now be running.
- take note that the public ip of the instance will change everytime you shut it down and launch it.
- You can pay to make it a static ip address.

## SSH-ing into your Instance
- You can use four ways of interacting with your server
- For example you can use a UNIX based system like Linux or Mac OS.
- You can even use Windows, but it requires you to download putty.
- You could also connect through your instance dashboard, but I will only go over using Linux and Windows.

## Using Linux
- Open up a terminal window

![Imgur](http://i.imgur.com/RA8r3Et.png)

- You will need to first chmod the key pair file you used for the server.
- go to where ever you put the key pair and type "chmod 400 insertNameofKeyPair" Note: do not type the quotes, and put in the name of the key pair instead of the insertNameofKeyPair.
- This will allow you to connect to the server. It would have given you a warning about the key pair being public or something similar.
- Type this in the terminal "ssh -i /path/my-key-pair-name.pem ubuntu@current-server-ip" Note: do not type the quotes, also change the command to point to the director of your pem file and change the ip to your servers current ip.

- If all goes well then you're logged in!

## Using Windows
- You will need to download and install [Putty](http://www.chiark.greenend.org.uk/~sgtatham/putty/download.html)
- choose the putty installer .exe under A Windows installer for everything except PuTTYtel.
- Putty needs to have a converted version of the pem file.
- open up Puttygen and load the pem file.

![Imgur](http://i.imgur.com/xoOgP3s.png)

- Note: make sure to change the dropdown list to all files when trying to select your pem file.
- After you have selected the pem file, you will need to convert it to a putty usable key file.
- Open up putty and put in "ubuntu@Current-ip-of-Server" into the area for host/ip text box.

![Imgur](http://i.imgur.com/B98Ew1T.png)

- On the left side, expand SSh and select Auth.
- Browse for the converted pem file.
- Click the Open Button to connect to the server.

- If all goes well then you're logged in!

## Making a webserver
- Now lets install some software to make the server into a webserver for hosting your files.
- type and enter each of the following commands
- sudo su
- apt-get update
- apt-get upgrade
- tasksel
- use arrow keys to go to lamp server and use spacebar to select it and use tab to go to OK and use enter.
- When it asks you to set a password for MySQl root user, set one and don't forget it.
- press tab to go to Ok and use enter
- congratulations you are technically done
- Now you have to use a UNIX based system, like linux or Mac OS, to send files using commands or on windows, use [WinSCP](https://winscp.net/eng/download.php) to send files to the server. Select installation pack to get the installer .exe for WinSCP.
- It is very similar to the way you connect with putty, and you can even import saved profiles made in putty as well.
